# O script classifica o vinho entre tinto ou branco

```python
import pandas as pd
arquivo = pd.read_csv("datasets_16721_22034_wine_dataset.csv")
```

```python
arquivo
```

### A primeira metade é de vinho tinto e a segunda metade é de vinho branco

```python
arquivo.head()
```

```python
arquivo.isnull().sum()
```

```python
# Red = tinto
arquivo['style'] = arquivo['style'].replace('red', 0)
```

```python
# white = branco
arquivo['style'] = arquivo['style'].replace('white', 1)
```

```python
arquivo.head()
```

```python
# Separando as variáveis entre preditoras e variáveis alvo
y = arquivo['style']
x = arquivo.drop('style', axis = 1)
```

```python
from sklearn.model_selection import train_test_split

# Criando os conjuntos de dados de treino e de teste
x_treino, x_teste, y_treino, y_teste, = train_test_split(x, y, test_size = 0.3)

# test_size = quantos % dos dados vão ser usados para teste? No caso foram 30%
```

### x é os dados que eu quero inputar para treinar e achar o y

```python
y_teste
```

```python
y_treino
```

### Acuracy do modelo

```python
# Esse é um algoritmo de classificar
from sklearn.ensemble import ExtraTreesClassifier

# Criando modelo 
modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino)

# Imprimindo resultados:
resultado = modelo.score(x_teste, y_teste)
print("Acurácia = ", resultado)
```
